# DAY - 2 : Visualizing Data (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Top 10 best-selling shooting games by global sales(PlayStation/XOne & year > 2013)
df = df.sort_values(by='Global_Sales', ascending=False)
platforms_of_interest = ['PS3', 'PS4', 'XOne', 'X360']
filt_new_shooter = df[(df["Year"] > 2013) & (df["Genre"] == "Shooter") & (df['Platform'].isin(platforms_of_interest))]
```

```python
filt_new_shooter.head(20)
```

```python
filt_new_shooter.value_counts()
```

```python
top_shooter_games = filt_new_shooter.nlargest(10, 'Global_Sales')
top_shooter_games
```

```python
# Create a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Global_Sales', y='Name', data=top_shooter_games, hue='Platform')
plt.title('Top 10 Best-Selling Shooter Games (Year > 2013) on PS3, PS4, XOne, X360')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Game Name')
plt.show()
```
